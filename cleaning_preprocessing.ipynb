```python


def replace_abr(address):
    
    address_abbreviations = {
    "AVENUE": "AVE",
    "BOULEVARD": "BLVD",
    "COURT": "CT",
    "CIRCLE": "CIR",
    "DRIVE": "DR",
    "LANE": "LN",
    "ROAD": "RD",
    "STREET": "ST",
    "SQUARE": "SQ",
    "TERRACE": "TER",
    "WAY": "WY",
    "PLACE": "PL",
    "ALLEY": "ALY",
    "PATH": "PTH",
    "PARKWAY": "PKWY",
    "VIA": "VIA",
    "PLAZA": "PLZ",
    "HILL": "HL",
    "CRESCENT": "CRES",
    "RANCH": "RCH",
    "SOUTH": "S",
    "NORTH": "N",
    "EAST": "E",
    "WEST": "W",
    "SOUTHWEST": "SW",
    "NORTHEAST": "NE",
    "SOUTHEAST": "SE",
    "NORTHWEST": "NW"
    }
    
    ending = address.split(' ')[-1]
    
    if ending in address_abbreviations:
        
        new_end = address_abbreviations[ending]
        
        parts = address.split(' ')[:-1]
        
        new_address = parts[0]
        
        for part in parts[1:]:
            
            new_address += f' {part}'
        
        address = f'{new_address} {new_end}'
        
        return address
        
    else:
        
        return address
```

```python
final_df['Address'] = final_df['property_street_address'].apply(replace_abr)
```
